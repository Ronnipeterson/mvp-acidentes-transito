**Descrição do Problema**

O objetivo deste projeto é analisar dados de acidentes de trânsito com vítimas ocorridos no ano de 2021, na cidade de Belo Horizonte e regiões próximas, com o intuito de identificar padrões temporais, geográficos e operacionais associados à ocorrência dos acidentes.

Trata-se de um problema de análise exploratória de dados (não supervisionado), uma vez que não há, neste momento, a construção de um modelo preditivo, mas sim a busca por compreensão dos dados e geração de insights relevantes.

Busca-se compreender os fatores que influenciam a ocorrência e a gravidade dos acidentes, especialmente no que diz respeito à presença de vítimas fatais, considerando variáveis como tipo de acidente, condições climáticas, tipo de pavimento, região e características da via.

Como premissa, assume-se que os dados disponibilizados representam adequadamente os registros oficiais de acidentes no período analisado, embora possam conter inconsistências, valores faltantes ou informações não informadas, que deverão ser tratados ao longo do processo de análise.

Este estudo possui como principal objetivo apoiar a geração de conhecimento que possa contribuir para a melhoria da segurança no trânsito, por meio da identificação de padrões e possíveis fatores de risco associados aos acidentes.


## Hipóteses do problema

Considerando o contexto dos acidentes de trânsito com vítimas em Belo Horizonte e regiões próximas, foram formuladas algumas hipóteses iniciais para orientar a análise exploratória dos dados. Essas hipóteses buscam direcionar a investigação de padrões temporais, espaciais e de gravidade dos acidentes.

- Os acidentes não se distribuem uniformemente ao longo do dia, concentrando-se em horários de maior circulação de veículos.

- A quantidade de acidentes varia ao longo da semana, com possível aumento em dias de maior mobilidade urbana, como a sexta-feira.

- Certos tipos de acidentes, como atropelamentos, tendem a apresentar maior gravidade e maior associação com fatalidades.

- A distribuição geográfica dos acidentes não é homogênea, havendo concentração em áreas urbanas mais movimentadas.

- Fatores contextuais, como velocidade permitida, tipo de pavimento e condição climática, podem influenciar a ocorrência e a gravidade dos acidentes.

## Tipo de problema

Este é um problema de classificação supervisionada. Dado um conjunto de características dos acidentes de trânsito, como tipo de acidente, horário, condições da via e região, o objetivo é analisar a variável alvo `INDICADOR_FATALIDADE`, que indica se houve ou não ocorrência de fatalidade.

```python

# Limpeza das coordenadas


df['COORDENADA_X'] = pd.to_numeric(df['COORDENADA_X'], errors='coerce')
df['COORDENADA_Y'] = pd.to_numeric(df['COORDENADA_Y'], errors='coerce')

# Remover registros com coordenadas ausentes ou zeradas
df_mapa = df.dropna(subset=['COORDENADA_X', 'COORDENADA_Y']).copy()
df_mapa = df_mapa[(df_mapa['COORDENADA_X'] != 0) & (df_mapa['COORDENADA_Y'] != 0)]

print("Registros válidos para mapa:", df_mapa.shape[0])
display(df_mapa[['COORDENADA_X', 'COORDENADA_Y']].head())
```

```python

# Conversão de coordenadas UTM -> latitude/longitude


!pip -q install pyproj

from pyproj import Transformer

# Hipótese: UTM zona 23S / SIRGAS 2000 ou WGS84
# EPSG:31983 = SIRGAS 2000 / UTM zone 23S
# EPSG:4326 = latitude/longitude

transformer = Transformer.from_crs("EPSG:31983", "EPSG:4326", always_xy=True)

df_mapa['LONGITUDE'], df_mapa['LATITUDE'] = transformer.transform(
    df_mapa['COORDENADA_X'].values,
    df_mapa['COORDENADA_Y'].values
)

display(df_mapa[['COORDENADA_X', 'COORDENADA_Y', 'LATITUDE', 'LONGITUDE']].head())
```

```python

# Verificação da faixa geográfica


print(df_mapa[['LATITUDE', 'LONGITUDE']].describe())
```

```python

# Mapa de acidentes


!pip -q install folium

import folium

centro_lat = df_mapa['LATITUDE'].mean()
centro_lon = df_mapa['LONGITUDE'].mean()

mapa = folium.Map(location=[centro_lat, centro_lon], zoom_start=11)

# Usei apenas uma amostra para não sobrecarregar o mapa
amostra_mapa = df_mapa.sample(min(1000, len(df_mapa)), random_state=42)

for _, row in amostra_mapa.iterrows():
    folium.CircleMarker(
        location=[row['LATITUDE'], row['LONGITUDE']],
        radius=2,
        fill=True
    ).add_to(mapa)

mapa
```

Além disso, a existência de áreas mais densamente afetadas pode auxiliar na identificação de pontos prioritários para intervenções de engenharia de tráfego, fiscalização e campanhas de prevenção. Dessa forma, a análise geográfica complementa a análise temporal e categórica, ampliando a compreensão do fenômeno estudado.

```python

# Tratamento de valores "não informado"


# Substituir por NaN
df.replace("NÃO INFORMADO", np.nan, inplace=True)
df.replace("NAO INFORMADO", np.nan, inplace=True)

# Verificação
df.isnull().sum().sort_values(ascending=False).head(10)
```

Nesta etapa, valores categóricos representados como "NÃO INFORMADO" foram convertidos para valores ausentes (NaN). Essa transformação é necessária para permitir o tratamento adequado de dados faltantes nas etapas subsequentes, evitando que esses valores sejam interpretados como categorias válidas.

```python

# Tratamento de valores ausentes


# Visualizar porcentagem de missing
missing_percent = df.isnull().mean().sort_values(ascending=False) * 100
display(missing_percent.head(10))
```

```python
# Remover colunas com muitos valores ausentes (>50%)
cols_drop = missing_percent[missing_percent > 50].index.tolist()
# Filter out columns that do not exist in the DataFrame
cols_to_drop_actual = [col for col in cols_drop if col in df.columns]
df.drop(columns=cols_to_drop_actual, inplace=True)

# Preencher categóricos com "Desconhecido"
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna("Desconhecido")

# Preencher numéricos com mediana
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    df[col] = df[col].fillna(df[col].median())
```

Os valores ausentes foram tratados de acordo com a natureza das variáveis. Colunas com alta proporção de dados faltantes foram removidas, por apresentarem baixo potencial informativo. Para variáveis categóricas, os valores ausentes foram substituídos por uma categoria genérica ("Desconhecido"), enquanto para variáveis numéricas foi utilizada a mediana, por ser uma medida robusta à presença de outliers.

```python

# Seleção de variáveis


cols_modelo = [
    'HORA',
    'DIA_SEMANA',
    'DESC_TIPO_ACIDENTE',
    'DESC_TEMPO',
    'PAVIMENTO', # Corrected from DESC_PAVIMENTO
    'DESC_REGIONAL', # Corrected from REGIAO
    'VELOCIDADE_PERMITIDA',
    'INDICADOR_FATALIDADE'
]

df_modelo = df[cols_modelo].copy()

df_modelo.head()
```

Foi realizada a seleção de variáveis relevantes para o problema, priorizando atributos com potencial explicativo sobre a ocorrência de fatalidades. Variáveis redundantes, identificadores e atributos sem variação foram descartados, a fim de reduzir a dimensionalidade e melhorar a interpretabilidade do conjunto de dados.

```python

# Encoding de variáveis categóricas


df_modelo = pd.get_dummies(df_modelo, drop_first=True)

df_modelo.head()
```

As variáveis categóricas foram transformadas em variáveis numéricas por meio da técnica de one-hot encoding, permitindo sua utilização em algoritmos de machine learning. A opção de remover a primeira categoria (drop_first=True) foi adotada para evitar multicolinearidade.

```python

# Ajuste da variável alvo


df_modelo['INDICADOR_FATALIDADE'] = df['INDICADOR_FATALIDADE'].map({
    'SIM': 1,
    'NÃO': 0,
    'NAO': 0
})
```

A variável alvo foi convertida para formato binário, sendo "SIM" representado por 1 (ocorrência de fatalidade) e "NÃO" por 0. Essa transformação é necessária para viabilizar a aplicação de modelos de classificação.

```python

# Dataset final


print("Dimensão final:", df_modelo.shape)
display(df_modelo.head())
```

Após as etapas de limpeza, tratamento e transformação dos dados, obteve-se um dataset estruturado, consistente e adequado para análise e modelagem. O conjunto final apresenta apenas variáveis relevantes, sem valores ausentes e em formato compatível com técnicas de machine learning.

```python
df_modelo['INDICADOR_FATALIDADE'].value_counts(normalize=True)
```

Observa-se que a variável alvo apresenta desbalanceamento entre as classes, o que pode impactar o desempenho de modelos preditivos. Esse aspecto deve ser considerado em etapas futuras de modelagem.
